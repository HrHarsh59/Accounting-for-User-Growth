```python
# Convert all values to strings and fill null values with empty strings
df = df.applymap(str).fillna('')

# Split the strings into lists of user IDs
df = df.applymap(lambda x: x.split() if x != '' else [])
users_by_week = {col: set(user for sublist in df[col] for user in sublist) for col in df.columns}
```

```python
new_users = []
retained_users = []
resurrected_users = []
churned_users = []
quick_ratios = []
```

```python
weeks = list(df.columns)
```

```python
weeks
```

```python
for i in range(1, len(weeks)):
    current_week = weeks[i]
    previous_week = weeks[i-1]
    
    current_users = users_by_week[current_week]
    previous_users = users_by_week[previous_week]
    
    new = len(current_users - previous_users)
    retained = len(current_users & previous_users)
    
    if i > 1:
        past_users = set().union(*[users_by_week[weeks[j]] for j in range(i-1)])
        resurrected = len((current_users & past_users) - previous_users)
    else:
        resurrected = 0  
    
    churned = len(previous_users - current_users)
    
    new_users.append(new)
    retained_users.append(retained)
    resurrected_users.append(resurrected)
    churned_users.append(churned)
    
    if churned != 0:
        quick_ratio = (new + resurrected) / churned
    else:
        quick_ratio = float('inf')  
    
    quick_ratios.append(quick_ratio)
```

```python
results = pd.DataFrame({
    'week' : weeks[1:],
    'new_users' : new_users,
    'retained_users' : retained_users,
    'resurrected_users' : resurrected_users,
    'churned_users' : churned_users,
    'quick_ratios' : quick_ratios
})
```

```python
results.head(52)
```

```python
plt.figure(figsize=(14, 8))
plt.bar(results['week'], results['new_users'], label='New Users', color='blue')
plt.bar(results['week'], results['retained_users'], bottom=results['new_users'], label='Retained Users', color='green')
plt.bar(results['week'], results['resurrected_users'], bottom=results['new_users'] + results['retained_users'], label='Resurrected Users', color='yellow')
plt.bar(results['week'], -results['churned_users'], label='Churned Users', color='red')
plt.xlabel('Week')
plt.ylabel('Number of Users')
plt.title('Weekly Growth Accounting')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

## KEY OBSERVATIONS

### New Users (Blue) :

- The number of new users remains relatively stable across weeks, indicating consistent user acquisition efforts.

- There are occasional spikes in new users, suggesting successful marketing campaigns or product launches during those weeks.

### Retained Users (Green) :

- The retained user base forms the largest component of active users, highlighting a solid core of loyal users.

- A noticeable increase in retained users over time suggests improving user satisfaction and engagement, which is a positive sign for long-term growth.


### Resurrected Users (Yellow) :

- The number of resurrected users fluctuates but shows an overall increasing trend.
    
- Effective re-engagement strategies, such as targeted marketing or feature updates, appear to be bringing previously inactive users back to the platform.

### Churned Users (Red) :

- The churn rate varies significantly week by week but remains substantial, indicating ongoing challenges with user retention.

- High churn in certain weeks may point to specific issues such as dissatisfaction with updates, bugs, or external factors affecting user behavior.
Strategic Insights

### `The user growth accounting metrics reveal a healthy user base with a strong core of retained users and a consistent influx of new users.
